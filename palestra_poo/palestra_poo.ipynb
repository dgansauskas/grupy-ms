<blockquote>
    <span style="font-family: Arial Bold Italic; font-size: 30px">Origem</span>
</blockquote>

<p style="font-family: Arial Italic; font-size: 18px">A tcnica de programao de computadores baseada em programao orientada a objetos no  recente. Seu reconhecimento pblico data praticamente do <span style="background: #FEF76E">inicio da dcada de 1990</span>, apesar da popularizao ter ocorrido a partir da dcada de 1970, no entanto, o conceito de programao orientada a objetos ou POO (Object Oriented Programming - OOP) surgiu por volta do inicio da dcada de 1960, a partir da necessidade de simulaes de computadores.</p>

<p style="font-family: Arial Italic; font-size: 18px">Em 1967 foi apresentada a linguagem <a href="https://pt.wikipedia.org/wiki/Simula_67"> SIMULA 67</a>, que introduziu o conceito de classes e blocos estruturados (procedimentos) como extenso da linguagem <a href="https://pt.wikipedia.org/wiki/ALGOL_60">ALGOL 60</a>, <span style="background: #FEF76E">sendo a classe um dos principais pilares que sustentam a programao orientada a objeto</span>. Durante a dcada de 1970, surge a linguagem de programao orientada a objetos <a href="https://pt.wikipedia.org/wiki/Smalltalk">SmallTalk</a> desenvolvida por Alan Kay. A SmallTalk de certa forma popularizou e incentivou o uso da POO, pois foi essa linguagem que introduziu o nome programao Orientada a Objetos. Durante a dcada de 80, algumas linguagens impulsionaram o paradigma da POO, destacando-se as linguagens: ADA, C++, Object PASCAL</p>

<blockquote>
    <span style="font-family: Arial Bold Italic; font-size: 30px">Definio</span>
</blockquote>    

<p style="font-family: Arial Italic; font-size: 18px">A orientao a objetos  um paradigma de anlise, projeto e programao de sistemas de software baseado na composio e interao entre diversas unidades de software chamadas objetos</p>

<p style="font-family: Arial Italic; font-size: 18px">A programao orientada a objetos  uma <span style="background: #FEF76E">filosofia de trabalho</span>. No  a ferramenta em si, mas a forma de pensar, a forma de usar a lgica de programao para a soluo de um problema do mundo real em um computador. Assim,  necessrio ao desenvolvedor mudar a forma de pensar um problema computacional.  preciso utilizar a estrutura de dados de um programa baseada na estrutura do mundo real</p>

<blockquote>
    <span style="font-family: Arial Bold Italic; font-size: 30px">Fundamentao</span>
</blockquote>

<p style="font-family: Arial Italic; font-size: 18px"> sabido que para manter um teto erguido, so necessrios, em mdia, quatro pilares-mestre. A programao orientada a objetos  fundamentada nessa estrutura, pois possui quatro pilares-mestre, que so: <span style="background: #FEF76E">classe, objeto, atributos e mtodos</span></p>

<blockquote>
    <blockquote>
        <span style="font-family: Arial Bold Italic; font-size: 30px">Classe</span>
    </blockquote>
</blockquote>

<p style="font-family: Arial Italic; font-size: 18px">O dicionrio Aurlio (2010) apresenta vrias definies para o termo classe, que pode ser coleo, grupo, conjunto de coisas afins, etc. D tambm a definio da rea de lgica da programao como: <span style="background: #FEF76E">classe  uma categoria descritiva geral, que abrange o conjunto de objetos que compartilham uma ou mais caractersticas quanto a seus  itens de dados e procedimentos associados.</span>  Em outras palavras, classe  um modelo usado para formatar a estrutura de um objeto, ou seja, uma estrutura usada para criar (instanciar) um objeto.</p>

```python
class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
        
    def __repr__(self):
        return f"Nome: {self.nome.title()}\nIdade: {self.idade}"
```

<p style="font-family: Arial Italic; font-size: 17px">Obs. nomes de classe devem ser escritos no estilo <a href="https://i.redd.it/6ooxjjkd204y.jpg">pascal case</a></p>

```python
p = Pessoa('luiz', 26)
print(p)
```

<p style="font-family: Arial Italic; font-size: 18px">A classe  a forma mais bsica de se <span style="background: #FEF76E">definir apenas uma nica vez como devem ser todos os objetos criados a partir dela</span>, em vez de definir cada objeto separadamente e at repetidamente.</p>

<p style="font-family: Arial Italic; font-size: 18px">Uma classe <span style="background: #FEF76E">pode ser derivada de outra classe existente.</span> Neste caso, chama-se <span style="background: #FEF76E">classe filho (ou sub-classe)</span>, enquanto a classe  existente denomina-se <span style="background: #FEF76E">classe pai (ou superclasse).</span> Desta forma  possvel determinar famlias de classes por meio de hierarquia. <span style="background: #FEF76E">A classe filho automaticamente herda os atributos e as funcionalidades da classe pai.</span> A este efeito d-se o nome de <span style="background: #FEF76E">herana (ou derivao).</span>  possvel tambm acrescentar atributos a uma classe filho, ou mesmo modificar os tributos herdados de uma classe pai. Quando isso ocorrer, utiliza-se especificao.</p>

```python
class Funcionario(Pessoa):
    def __init__(self, nome, idade, matricula):
        super().__init__(nome, idade)
        self.matricula = matricula
    
    def __repr__(self):
        return super().__repr__() + f'\nMatrcula: {self.matricula}'
```

```python
f = Funcionario('Vagner Vinicios', 27, 654323)
print(f)
```

<span style="font-family: Arial Bold Italic; font-size: 25px">Associao</span>

<p style="font-family: Arial Italic; font-size: 18px">A herana  til quando precisamos obter um reso de membros e, principalmente, uma definio de subtipos. Embora estas situaes sejam comuns e teis em aplicaes orientadas a objetos, no  a nica necessidade de relacionamento.</p>

<p style="font-family: Arial Italic; font-size: 18px">Como j foi visto antes, a herana tem como finalidade possibilitar a criao de subtipos mais do que reutiliz-los.</p>

<p style="font-family: Arial Italic; font-size: 18px">Ento, na hora de criar um relacionamento atravs da herena, cabe uma pergunta: Um objeto  outro? Se a resposta for no, voc dever buscar relacionar uma classe a outra atravs da associao.</p>

<p style="font-family: Arial Italic; font-size: 18px">Associao possibilita um relacionamento entre classes/objetos, no qual estes possam pedir ajuda a outros e assim representar de forma completa o conceito no qual se destinam. Neste tipo de relacionamento, as classes e os objetos interagem entre si para atingir seus objetivos.<br><cite>Thiago Leite E Carvalho.</cite> <cite>Orientao a Objetos</cite></p>

<span style="font-family: Arial Bold Italic; font-size: 20px">Agregao</span>

```python
class Endereco:
    def __init__(self, rua, numero):
        self.rua = rua
        self.numero = numero
        
    def __repr__(self):
        return f"Endereo\nRua: {self.rua}\nN: {self.numero}"

class Funcionario(Pessoa):
    def __init__(self, nome, idade, matricula, endereco):
        super().__init__(nome, idade)
        self.matricula = matricula
        self.endereco = endereco
    
    def __repr__(self):
        return (
            super().__repr__() + f"\nMatrcula: {self.matricula}"
            + "\n" + self.endereco.__repr__()
        )
```

```python
e = Endereco('jos mauro vasconcelos', 100)
f = Funcionario('marcos', 31, 543223, e)
print(f)
```

<span style="font-family: Arial Bold Italic; font-size: 20px">Composio</span>

```python
class Endereco:
    def __init__(self, rua, numero):
        self.rua = rua
        self.numero = numero
        
    def __repr__(self):
        return f"Endereo\nRua: {self.rua}\nN: {self.numero}"

class Funcionario(Pessoa):
    def __init__(self, nome, idade, matricula, rua, numero):
        super().__init__(nome, idade)
        self.matricula = matricula
        self.endereco = Endereco(rua, numero)
    
    def __repr__(self):
        return (
            super().__repr__() + f"\nMatrcula: {self.matricula}"
            + "\n" + self.endereco.__repr__()
        )
```

```python
f = Funcionario('marcos', 31, 543223, 'jos mauro vasconcelos', 100)
print(f)
```

<p style="font-family: Arial Italic; font-size: 18px">Mtodo  uma poro de cdigo  que  disponibilizada pela classe. Este  executado quando  feita uma requisio a ele. <span style="background: #FEF76E">Um mtodo serve para identificar quais servios, aes, que a classe oferece.</span> Eles so responsveis por definir e realizar um determinado comportamento.</p>

```python
class Pessoa:
    def __init__(self, nome, idade, sobrenome=None):
        self.nome = nome
        self.sobrenome = sobrenome
        self.idade = self.verifica_idade(idade)
        
    def nome_completo(self):
        if self.sobrenome is not None:
            return f"{self.nome.title()} {self.sobrenome.title()}"
        else:
            return self.nome.title()
        
    def verifica_idade(self, idade):
        if idade < 18:
            raise Exception('No  permitido pessoas menores de idade')
        else:
            return idade
        
    def __repr__(self):
        return f"Nome: {self.nome_completo()}\nIdade: {self.idade}"
```

<p style="font-family: Arial Italic; font-size: 18px">Atributos  e mtodos podem ser de dois tipos, de instncia ou esttico.</p>

<p style="font-family: Arial Italic; font-size: 18px">Os atributos de instncia, embora definidos na classe, <span style="background: #FEF76E">pertencem ao objeto.</span> Ou seja, s podero ser acessados/utilizados a partir de instncias de uma classe, no caso um objeto. Com isso, cada um pode ter valores distintos para seus atributos e assim conseguir armazenar os valores que necessitam.</p>

```python
class Math:
    @staticmethod
    def soma(i, j):
        print("Resultado: " + str(i + j))
```

```python
class Pessoa:
    def __init__(self, nome, idade, sobrenome=None):
        self.nome = nome
        self.sobrenome = sobrenome
        self.idade = self.verifica_idade(idade)
        
    def nome_completo(self):
        if self.sobrenome is not None:
            return f"{self.nome.title()} {self.sobrenome.title()}"
        else:
            return self.nome.title()
        
    def verifica_idade(self, idade):
        if idade < 18:
            raise Exception('No  permitido pessoas menores de idade')
        else:
            return idade
        
    def __repr__(self):
        return f"Nome: {self.nome_completo()}\nIdade: {self.idade}"
```

```python
class Conta(Pessoa):
    agencia = 5543
    def __init__(self, nome, idade, sobrenome):
        super().__init__(nome, idade, sobrenome)
        self.__saldo = 0.0
        
    @property
    def saldo(self):
        return self.__saldo
    
    def saca(self, valor):
        if valor <= self.__saldo:
            self.__saldo -= valor
            return valor
        else:
            raise Exception('Saldo indisponivel!!!')
        
    def deposita(self, valor):
        self.__saldo += valor
    
    def __repr__(self):
        return super().__repr__() + f"\nSaldo: {self.saldo}"
    
    @classmethod
    def retorna_agencia(cls):
        return cls.agencia
```

```python
c = Conta('luiz', 26, 'henrique')
```

```python
print(c)
```

```python
c.agencia
```

```python
c2 = Conta('vagner', 27, 'vinicios')
```

```python
print(c2)
```

```python
c.agencia
```

```python
Conta.retorna_agencia()
```
